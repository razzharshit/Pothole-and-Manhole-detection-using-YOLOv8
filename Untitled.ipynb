```python
import os

# Define paths for images and labels
image_folder = r'C:\Users\rajha\Downloads\archive (1)\dataset\dataset\valid\images'
label_folder = r'C:\Users\rajha\Downloads\archive (1)\dataset\dataset\valid\labels'

# List the image and label files (filter by extensions)
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]
label_files = [f for f in os.listdir(label_folder) if f.endswith('.txt')]

# Extract base names without extensions for comparison
image_basenames = {os.path.splitext(file)[0] for file in image_files}
label_basenames = {os.path.splitext(file)[0] for file in label_files}

# Find the images that don't have corresponding labels
unmatched_images = image_basenames - label_basenames

# Delete unmatched images
for image_name in unmatched_images:
    # Construct the full file path
    for extension in ['.jpg', '.png', '.jpeg']:
        image_path = os.path.join(image_folder, image_name + extension)
        if os.path.exists(image_path):
            os.remove(image_path)
            print(f"Deleted unmatched image: {image_path}")

# Print completion message
print("Unmatched images deleted successfully.")
```

```python
import os
import cv2
import albumentations as A
from tqdm import tqdm

# Local paths
images_path = r'C:\Users\rajha\Downloads\archive (1)\dataset\dataset\classes\cracks\images'
labels_path = r'C:\Users\rajha\Downloads\archive (1)\dataset\dataset\classes\cracks\labels'
output_images_path = r'C:\Users\rajha\Downloads\augmented\images'
output_labels_path = r'C:\Users\rajha\Downloads\augmented\labels'

# Create output directories if they don't exist
os.makedirs(output_images_path, exist_ok=True)
os.makedirs(output_labels_path, exist_ok=True)

# Augmentation pipeline
augment = A.Compose([
    A.Rotate(limit=30, border_mode=0, p=1.0),
    A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=1.0),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
], bbox_params=A.BboxParams(format='yolo', label_fields=['category_ids']))

# Get valid image-label file pairs
image_files = sorted([f for f in os.listdir(images_path) if f.endswith(('.jpg', '.png'))])
label_files = sorted([f for f in os.listdir(labels_path) if f.endswith('.txt')])

# Ensure matching filenames
paired_files = [(img, img.replace('.jpg', '.txt').replace('.png', '.txt')) 
                for img in image_files if img.replace('.jpg', '.txt').replace('.png', '.txt') in label_files]

current_image_count = len(paired_files)
target_image_count = 500
new_image_id = current_image_count + 1

for image_file, label_file in tqdm(paired_files):
    if current_image_count >= target_image_count:
        break

    image_path = os.path.join(images_path, image_file)
    label_path = os.path.join(labels_path, label_file)
    image = cv2.imread(image_path)

    if image is None:
        print(f"Skipping invalid image: {image_file}")
        continue

    # Read labels
    bboxes = []
    category_ids = []
    with open(label_path, 'r') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) != 5:
                print(f"Skipping invalid line in {label_file}: {line}")
                continue
            try:
                class_id, x_center, y_center, width, height = map(float, parts)
                bboxes.append([x_center, y_center, width, height])
                category_ids.append(int(class_id))
            except ValueError:
                print(f"Corrupt values in {label_file}: {line}")
                continue

    if not bboxes:
        continue

    # Apply augmentation
    augmented = augment(image=image, bboxes=bboxes, category_ids=category_ids)

    # Save image and label
    new_img_name = f'{new_image_id:04}.jpg'
    cv2.imwrite(os.path.join(output_images_path, new_img_name), augmented['image'])

    new_lbl_name = f'{new_image_id:04}.txt'
    with open(os.path.join(output_labels_path, new_lbl_name), 'w') as f:
        for bbox, category_id in zip(augmented['bboxes'], augmented['category_ids']):
            f.write(f"{category_id} {' '.join(map(str, bbox))}\n")

    new_image_id += 1
    current_image_count += 1

print(f"âœ… Augmentation completed! Total images: {current_image_count}")
```

```python
import os
import cv2
import albumentations as A
from tqdm import tqdm

# Paths
images_path = r'C:\Users\rajha\Downloads\archive (1)\dataset\dataset\classes\cracks\images'
labels_path = r'C:\Users\rajha\Downloads\archive (1)\dataset\dataset\classes\cracks\labels'
output_images_path = r'C:\Users\rajha\Downloads\archive (1)\dataset\dataset\classes\cracks\augmented\images'
output_labels_path = r'C:\Users\rajha\Downloads\archive (1)\dataset\dataset\classes\cracks\augmented\labels'

# Create output directories
os.makedirs(output_images_path, exist_ok=True)
os.makedirs(output_labels_path, exist_ok=True)

# Augmentations
augmentations = {
    "rotate": A.Compose([A.Rotate(limit=30, border_mode=0, p=1.0)], bbox_params=A.BboxParams(format='yolo', label_fields=['category_ids'])),
    "hflip": A.Compose([A.HorizontalFlip(p=1.0)], bbox_params=A.BboxParams(format='yolo', label_fields=['category_ids'])),
    "vflip": A.Compose([A.VerticalFlip(p=1.0)], bbox_params=A.BboxParams(format='yolo', label_fields=['category_ids'])),
    "brightness": A.Compose([A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=1.0)], bbox_params=A.BboxParams(format='yolo', label_fields=['category_ids'])),
}

# Files
label_files = sorted([f for f in os.listdir(labels_path) if f.endswith('.txt') and os.path.isfile(os.path.join(labels_path, f))])
image_files = sorted([f for f in os.listdir(images_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))])


current_image_count = 667
max_image_count = 1000
new_image_id = current_image_count

for image_file, label_file in tqdm(zip(image_files, label_files), total=len(image_files)):
    if new_image_id > max_image_count:
        break

    image_path = os.path.join(images_path, image_file)
    label_path = os.path.join(labels_path, label_file)
    image = cv2.imread(image_path)

    # Read YOLO labels
    with open(label_path, 'r') as f:
        lines = f.readlines()

    bboxes = []
    category_ids = []
    for line in lines:
        try:
            class_id, x_center, y_center, width, height = map(float, line.strip().split())
            bboxes.append([x_center, y_center, width, height])
            category_ids.append(int(class_id))
        except ValueError:
            print(f"Invalid line in {label_file}: {line}")
            continue

    if not bboxes:
        continue

    for name, augmenter in augmentations.items():
        if new_image_id > max_image_count:
            break

        augmented = augmenter(image=image, bboxes=bboxes, category_ids=category_ids)

        # Save image
        new_img_name = f'{new_image_id:04}.jpg'
        cv2.imwrite(os.path.join(output_images_path, new_img_name), augmented['image'])

        # Save label
        new_lbl_name = f'{new_image_id:04}.txt'
        with open(os.path.join(output_labels_path, new_lbl_name), 'w') as f:
            for bbox, category_id in zip(augmented['bboxes'], augmented['category_ids']):
                f.write(f"{category_id} {' '.join(map(str, bbox))}\n")

        new_image_id += 1

print(f"âœ… Augmentation completed! Total new images: {new_image_id - current_image_count}")
```
